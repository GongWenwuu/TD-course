```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

## Missing Scenario

#### MAR (Missing at random)

```python
image = imageio.imread('datasets/color-images/{}.bmp'.format("lena"))
dim1, dim2, dim3 = image.shape
rate = 0.3 # Set missing rate
mask = np.round(np.random.rand(dim1, dim2) + 0.5 - rate)  # Generate a binary mask.
image0 = image.copy()
for k in range(dim3):
    image0[:, :, k] = np.multiply(image[:, :, k], mask) # Generate incomplete image
sparse_image = image0.copy()        
pos = np.where((image != 0) & (sparse_image == 0))
print('Image: %s' % ("lena"))
print('Missing rate: {:.2}'.format(rate))

plt.imshow(sparse_image)
plt.axis('off')
plt.savefig("results\lena_mar.eps")
plt.show()
```

#### R-MAR (Row-wise missing at random)

```python
image = imageio.imread('datasets/color-images{}.bmp'.format("lena"))
dim1, dim2, dim3 = image.shape
rate = 0.3 # Set missing rate
mask_row = np.round(np.random.rand(dim1) + 0.5 - rate)  # Generate a binary mask.
image0 = image.copy()
for i in range(dim2):
    for j in range(dim3):
        image0[:, i, j] = np.multiply(image[:, i, j], mask_row)
sparse_image = image0.copy() # Generate incomplete image      
pos = np.where((image != 0) & (sparse_image == 0))
print('Image: %s' % ("lena"))
print('Missing rate: {:.2}'.format(rate))

plt.imshow(sparse_image)
plt.axis('off')
plt.show()
```

#### C-MAR (Column-wise missing at random)

```python
image = imageio.imread('data/{}.bmp'.format("lena"))
dim1, dim2, dim3 = image.shape
missing = 0
rate = 0.1 * (2 * missing + 3) # Set missing rate
mask_column = np.round(np.random.rand(dim2) + 0.5 - rate)  # Generate a binary mask.
image0 = image.copy()
for i in range(dim2):
    for j in range(dim3):
        image0[i, :, j] = np.multiply(image[i, :, j], mask_column)
sparse_image = image0.copy() # Generate incomplete image      
pos = np.where((image != 0) & (sparse_image == 0))
print('Image: %s' % ("lena"))
print('Missing rate: {:.2}'.format(rate))

plt.imshow(sparse_image)
plt.axis('off')
plt.show()
```

#### RC-MAR ((Row, column)-wise missing at random)

```python
image = imageio.imread('data/{}.bmp'.format("lena"))
dim1, dim2, dim3 = image.shape
rate = 0.3 # Set missing rate
mask_row = np.round(np.random.rand(dim1) + 0.5 - (1 - np.sqrt(1 - rate)))
mask_column = np.round(np.random.rand(dim2) + 0.5 - (1 - np.sqrt(1 - rate)))
image0 = image.copy()
for i in range(dim2):
    for j in range(dim3):
        image0[:, i, j] = np.multiply(image[:, i, j], mask_row)
sparse_image = image0.copy() # Generate incomplete image
for i in range(dim1):
    for j in range(dim3):
        sparse_image[i, :, j] = np.multiply(image0[i, :, j], mask_column)
pos = np.where((image != 0) & (sparse_image == 0))
print('Image: %s' % ("lena"))
print('Missing rate: {:.2}'.format(rate))

plt.imshow(sparse_image)
plt.axis('off')
plt.show()
```
